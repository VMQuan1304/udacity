## Investigation Overview

My interest topic is medical fee billed by health insurance, which elements can impact on this variable. To clear this point, I analyze in by python (seaborn, matplitlib) to  visualize the dataset. After investigation, smokers, sex are variables that impact on charges variable. To be specifc, the responsers who do not smoke intend to invest more money on health insurance than smokers. Besides, the number of male who pay insurance fee more than male. Age, children, bmi variables do not affect on health insurance much.



## Dataset Overview

This is Health insurance dataset containing 1338 rows and 7 variables including age of responsers, sex, bmi (body mass index), children, smoker, region and charges. After checking step, the dataset has no null values and outliers. To deep dive each variables, please follow the explaination below:

* age: age of primary beneficiary
* sex: insurance contractor gender, female, male
* bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height, objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9
* children: Number of children covered by health insurance / Number of dependents
* smoker: Smoking
* region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.
* charges: Individual medical costs billed by health insurance


````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import log
import warnings
%matplotlib inline
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Health_insurance.csv')
df.head()
````

## How is density of charges variable?

>Charges variable has no the same scale with the orthers. For example, this variable have scale around 10000-30000, while age has range 20-50. That will affect on regression or prediction after descrption so I need to re-scale it to range 0-1 by logarit function. Simultaneously, the distribution of this variable change from skewed distribution to normal distribution.

````python
# visualize density of charges variable
fig =plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
ax=sns.distplot(df.charges);

ax1 = fig.add_subplot(122)
ax1=sns.distplot(log(df.charges));

# Set common labels
ax1.set_xlabel('Logarit of charges')
ax1.set_ylabel('Density')
ax1.set_title('Logarit of charges')

ax.set_title('Density of Charges')
ax.set_xlabel('Charges')
````

## What percentage is smokers in dataset?

> The percentage of smokers is fourtimes larger than ones who do not smoke. 

````python
# visualize the percentage of smokers in dataset
colors = ['#66b3ff', '#ffcc99']
explode = (0.05, 0.05)
df.groupby(['smoker']).count().plot(kind='pie',
                                 y='age', 
                                 autopct='%.0f%%', 
                                 explode=explode, 
                                 figsize=(8,8),
                                 colors=colors,
                                 legend=True, title='Smoker',
                                 fontsize=14)
plt.axis('off');
````

## How is children distribution?

> Responsers with no children are the highest bar.The families with 4 or 5 children are the lowest bar.

````python
# visualize chidren distribution
fig =plt.figure(figsize=(15,5))
sns.histplot(df.children);
plt.title("Children distribution", fontweight='bold');
plt.xlabel("Number of children");
plt.ylabel("Frequency");
````

## How is Charge distribution by smokers?

> Smokers variable impacts on charges variable. To be specifc, the smokers intend to have medical costs billed by health insurance than responsers who do not smoke.

````python
# visualize charge distribution by smokers
fig =plt.figure(figsize=(15,5))
sns.histplot(data=df, x=df['charges'],  hue='smoker',  multiple="stack", palette="Blues_r")
plt.title("Charge distribution by smokers", fontweight='bold');
plt.xlabel("Charges");
plt.ylabel("Smokers");
````

## How are Children and charges by smoker?

> Responsers who have less children in insurance health program will have more medical fee paid by insurance health, especially smokers. That can prove smokers with no children can have high medical fee paid by insurance health. Moreover, responsers have more children, the less smokers we have.

````python
# visualize Children and charges by smoker
fig =plt.figure(figsize=(15,5))
sns.stripplot(data=df, x="children", y="charges", hue="smoker")
plt.title("Children and charges by smoker", fontweight='bold');
plt.xlabel("Children");
plt.ylabel("Charges");
````

## Conclusions

> In conclusion, responsers who have less children in insurance health program will have more medical fee paid by insurance health, especially smokers. Moreover, sex of smokers and non-smokers does not impact much on medical fee paid by insurance health. Sex and smoking behavior do not affect on bmi, which is Body mass index, providing an understanding of body, weights. Moreover, responsers have more children, the less smokers we have.
